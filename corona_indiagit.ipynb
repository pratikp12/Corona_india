```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
import seaborn as sns
covid=pd.read_csv("covid_19_india.csv")
covid
```

# Removing unwanted colums

```python
covid1=covid.drop(['ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
covid1
```

# selecting rows which show data of date 02/05/20 

```python
covid_latest=covid1[covid1['Date']=='02/05/20']
```

# Performing data processing

```python
#sorting rows based on Confirmed cases statewise
covid_latest.sort_values('Confirmed')
```

```python
#sorting rows based on Confirmed cases statewise in descending order date 02/05/20
#showing first 10 rows 
covid_latest.sort_values(by='Confirmed', ascending=False)[:10]
```

```python
#sorting rows based on Death statewise in descending order date 02/05/20
#showing first 10 rows
# here you can see Maharashtra state Death number is high
covid_latest.sort_values(by='Deaths', ascending=False)[:10]
```

```python
#Bar graph that Top 10 state with Confirmed case 
covid_latest.sort_values(by='Confirmed', ascending=False)[:10].plot(x='State/UnionTerritory',y='Confirmed',kind='bar')
```

```python
#graph showing number of cases in Maharashtra state increase with respect to date
mh=covid[covid['State/UnionTerritory']=='Maharashtra'].plot(x='Date',y='Confirmed',figsize = (10,6),fontsize=16)
mh.legend_.remove()

mh.set_title("Maharashtra", fontsize=20)
mh.set_xlabel('Dates', fontsize=20)
mh.set_ylabel('No. of cases', fontsize=20)
```

```python
#graph showing Goa not having new cases in april month
goa=covid[covid['State/UnionTerritory']=='Goa'].plot(x='Date',y='Confirmed',figsize = (10,6),fontsize=16)
goa.legend_.remove()
goa.set_title("Goa", fontsize=20)
goa.set_xlabel('Dates', fontsize=20)
goa.set_ylabel('No. of cases', fontsize=20)
```

```python
#Manipur state work well in fight against corona
manipur=covid[covid['State/UnionTerritory']=='Manipur'].plot(x='Date',y='Confirmed',figsize = (10,6),fontsize=16)
manipur.legend_.remove()
manipur.set_title("Manipur", fontsize=20)
manipur.set_xlabel('Dates', fontsize=20)
manipur.set_ylabel('No. of cases', fontsize=20)
```
